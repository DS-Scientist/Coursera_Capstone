# Scrape the list of Postal Codes of Canada along with Borough and Neighbourhood.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
wiki_url = requests.get(url)
wiki_url
```

```python
wiki_data = pd.read_html(wiki_url.text)
wiki_data
```

# Data Cleansing: One Postal Code can have multiple Neighbourhoods.

# Data Cleansing: Determine Postal Codes along with Boroughs & Neighborhoods that are 'Not assigned'.

# Data Cleansing: If a Borough has a 'Not assigned' Neighbourhood, then the Neighbourhood is the same as the Borough.

```python
len(wiki_data), type(wiki_data)
```

```python
wiki_data = wiki_data[0]
wiki_data
```

# Data Cleansing: Only rows with an assigned borough are processed; if 'Not assigned', these are ignored (removed) from the list.

```python
df = wiki_data[wiki_data["Borough"] != "Not assigned"]
df
```

# Data Cleansing: Postal Codes can have multiple Neighbourhoods; the 'Not assigned' ones have been removed.

```python
df = df.groupby(['Postal Code']).head()
df
```

```python
df.Neighbourhood.str.count("Not assigned").sum()
```

```python
df = df.reset_index()
df
```

```python
df.drop(['index'], axis = 'columns', inplace = True)
df
```
